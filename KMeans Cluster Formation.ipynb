```python
#Importing Required
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
#Reading the dataset and omitting the not-required columns
df=pd.read_csv("/content/Iris.csv")
df
```

```python
iris=df.drop(['Id', 'Species'], axis=1)
iris
```

```python
#Converting the 'columns-omitted-dataset' into the  integer-location based indexing form.
#.values is only returning the values of the data frame with the axis labels removed.
#.iloc uses integer-location based indexing.
x = iris.iloc[:, 0:4].values
x
```

```python
# Finding the optimum number of clusters for k-means classification
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)


plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#The optimum value of K (elbow) is found to be 3 by oserving the graph
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
#Scatterplot formation
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],
            s = 50, c = 'r', marker =",", label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],
            s = 50, c = 'b', marker ="v", label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 50, c = 'g', marker =">", label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],
            s = 200, c = 'y', marker ="o", label = 'Centroids')

plt.legend()
```
